```python
# Data handling aur preprocessing ke liye
import pandas as pd
import numpy as np
import os
```

```python
# Dataset ko dubara load kar rahe hain
# Taaki notebook independent rahe

file_path = "/content/drive/MyDrive/Customer_churn_Telecom/data/raw/Telco_customer_churn.xlsx"
df = pd.read_excel(file_path)

df.head()
```

```python
# Cleaning se pehle dataset ka size
df.shape
```

```python
# Total Charges ko numeric banate hain
df["Total Charges"] = pd.to_numeric(df["Total Charges"], errors="coerce")
```

```python
# Sirf important columns ke NaN hata rahe hain
df = df.dropna(subset=["Total Charges"])
```

```python
selected_cols = [
    "Tenure Months",
    "Monthly Charges",
    "Senior Citizen",
    "Contract",
    "Internet Service",
    "Payment Method",
    "Churn Value"     # ðŸ‘ˆ TARGET (MUST)
]

df = df[selected_cols]

df.head()
```

```python
# Senior Citizen ko binary banate hain
df["Senior Citizen"] = df["Senior Citizen"].map({"Yes": 1, "No": 0})
```

```python
df["Senior Citizen"].value_counts(dropna=False)
```

```python
categorical_cols = [
    "Contract",
    "Internet Service",
    "Payment Method"
]

df = pd.get_dummies(df, columns=categorical_cols)
```

```python
df.to_csv("/content/drive/MyDrive/Customer_churn_Telecom/data/processed/churn_cleaned.csv", index=False)
```
